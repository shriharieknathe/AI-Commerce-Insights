```python
import mysql.connector
import pandas as pd

# Establish connection to MySQL database
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mindstix",
    database="edb"
)

# Cursor to fetch tables
cursor = connection.cursor()

# Query to get all table names
cursor.execute("SHOW TABLES")

# Get a list of all table names
tables = [table[0] for table in cursor.fetchall()]

# Create an empty list to store DataFrames
dataframes = []

# Iterate through all tables and load each into a DataFrame
for table in tables:
    query = f"SELECT * FROM {table}"
    df = pd.read_sql(query, con=connection)
    dataframes.append(df)

# Close the connection
connection.close()

# Now you have a list of DataFrames
for i, df in enumerate(dataframes):
    print(f"DataFrame for table {tables[i]}:\n", df.head())  # Print the first few rows of each DataFrame
```
